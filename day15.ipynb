```python
strings = open("./data/input_day15.txt").read().strip().split(",")
strings
```

```python
from string import ascii_letters

def hash_function(s):
    h = 0
    for c in s:
        h += ord(c)
        h *= 17
        h %= 256
    return h

hash_function("HASH")
```

```python
total = 0
for s in strings:
    total += hash_function(s)

total
```

```python
from collections import OrderedDict
import re

boxes = {b: OrderedDict() for b in range(256)}
for s in strings:  
    lens = re.search(r"[a-z]+", s).group(0)
    box = hash_function(lens)
    if "=" in s:
        strength = s.split("=")[-1]
        boxes[box][lens] = int(strength)
    if "-" in s:
        if lens in boxes[box]:
            del boxes[box][lens]

boxes = {b: v for b, v in boxes.items() if v}
total = 0
for box, lenses in boxes.items():
    for i, strength in enumerate(lenses.values(), start=1):
        total += (box + 1) * i * strength
total
```
