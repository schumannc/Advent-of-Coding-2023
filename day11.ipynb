```python
image = open("./data/input_day11.txt").read().splitlines()
image = [list(line) for line in image]
image[0][:5]
```

```python
traspose_image = [list(l) for l in list(zip(*image))]
traspose_image[0][:5]
```

```python
rows_to_expand = []
for i, row in enumerate(image):
    if set(row) == {"."}:
        rows_to_expand.append(i)

cols_to_expand = []
for i, col in enumerate(traspose_image):
    if set(col) == {"."}:
        cols_to_expand.append(i)
```

```python
galaxies = []
for row in range(len(image)):
    for col in range(len(image[0])):
        if image[row][col] == "#":
            galaxies.append((row, col))
```

```python
new_galaxies = []
for g in galaxies:
    mi = 1_000_000 - 1
    rows_before = [r for r in rows_to_expand if r < g[0]]
    columns_before = [c for c in cols_to_expand if c < g[1]]
    new_galaxies.append((g[0] + len(rows_before) * mi, g[1] + len(columns_before) * mi))

new_galaxies
```

```python
total = 0
combinations = 0
for i, g in enumerate(new_galaxies):
    for j, g2 in enumerate(new_galaxies[i+1:]):
        combinations += 1
        shotest = abs(g[0] - g2[0]) + abs(g[1] - g2[1])
        total += shotest

total, combinations
```
