```python
digplan = open("./data/input_day18.txt").read().splitlines()
digplan = list(map(lambda x: x.split(), digplan))
digplan
```

```python
from collections import deque
row, col = 0, 0
moves = {'L': (0, -1), 'R': (0, 1), 'U': (-1, 0), 'D': (1, 0)} 
lagoon = []
for d, s, _ in digplan:
    dy, dx = moves[d]
    s = int(s)
    if d in ('L', 'R'):
        for _ in range(s):
            col += dx
            lagoon.append((row, col))
    else:
        for _ in range(s):
            row += dy
            lagoon.append((row, col))

min_col = min([c for _, c in lagoon])
min_row = min([r for r, _ in lagoon])
for i in range(len(lagoon)):
    r, c = lagoon[i]
    lagoon[i] = (r - min_row, c - min_col)
    
max_col = max([c for _, c in lagoon])
max_row = max([r for r, _ in lagoon])
grid = [["."] * (max_col + 1) for _ in range(max_row + 1)]
for r, c in lagoon:
    grid[r][c] = "#"


def find_flat(i, j, grid, seen):
    start = (i, j)
    frontier = deque([start])
    seen.add(start)
    while len(frontier) > 0:
        i, j = frontier.popleft()
        for di, dj in [(0, 1), (0, -1), (1, 0), (-1, 0)]:
            if 0 <= i + di < len(grid) and 0 <= j + dj < len(grid[0]):
                if grid[i + di][j + dj] == "." and (i + di, j + dj) not in seen:
                    seen.add((i + di, j + dj))
                    frontier.append((i + di, j + dj))
    return seen



# flatland = set()
# for i in [0, max_row]:
#     for j in range(max_col + 1):
#         c = grid[i][j]
#         k = i
#         while c == ".":
#             flatland.add((k, j))
#             k += 1 if i == 0 else -1
#             c = grid[k][j]


# for j in [0, max_col]:
#     for i in range(max_row + 1):
#         c = grid[i][j]
#         k = j
#         while c == ".":
#             flatland.add((i, k))
#             k += 1 if j == 0 else -1
#             c = grid[i][k]
        
flatland = set()
for i in [0, max_row]:
    for j in range(max_col + 1):
        c = grid[i][j]
        if c == "." and (i, j) not in flatland:
            seen = find_flat(i, j, grid, flatland)
            flatland |= seen

for j in [0, max_col]:
    for i in range(max_row + 1):
        c = grid[i][j]
        if c == "." and (i, j) not in flatland:
            seen = find_flat(i, j, grid, flatland)
            flatland |= seen
            
for i in range(max_row + 1):
    for j in range(max_col + 1):
        if grid[i][j] == "." and (i, j) not in flatland:
            grid[i][j] = "#"
           
grid
```
