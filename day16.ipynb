```python
contraption = open('./data/input_day16.txt').read().splitlines()
contraption = list(map(list, contraption))
contraption
```

```python
def print_contraption(contraption):
    for row in contraption:
        for col in row:
            s = len(col)
            print(col + " " * (2 - s), end='')
        print()

print_contraption(contraption)
```

```python
from collections import deque

direction = (0, 1)
tiles = []

splits = deque()
row, col = 0, 0
while True:
    if row < 0 or row >= len(contraption) or col < 0 or col >= len(contraption[0]) or (row, col, direction) in tiles:
        if len(splits) == 0:
            break
        else:
            row, col, direction = splits.popleft()
    
    tile = contraption[row][col]
    tiles.append((row, col, direction))
    
    if tile == "|":
        if direction in [(0, 1), (0, -1)]:
            if row - 1 >= 0:
                splits.append((row, col, (-1, 0)))
            direction = (1, 0)
       
    elif tile == "-":
        if direction in [(1, 0), (-1, 0)]:
            if col - 1 >= 0:
                splits.append((row, col, (0, -1)))
            direction = (0, 1)
       
    elif tile == "/":
        if direction == (0, 1):
            direction = (-1, 0)
        elif direction == (0, -1):
            direction = (1, 0)
        elif direction == (1, 0):
            direction = (0, -1) 
        elif direction == (-1, 0):
            direction = (0, 1)
            
    elif tile == "\\":
        if direction == (0, 1):
            direction = (1, 0)
        elif direction == (0, -1):
            direction = (-1, 0)
        elif direction == (1, 0):
            direction = (0, 1)
        elif direction == (-1, 0):
            direction = (0, -1)

    row, col = row + direction[0], col + direction[1]
```
