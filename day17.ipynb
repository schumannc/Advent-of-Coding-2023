```python
from dataclasses import dataclass

@dataclass
class Point:
    parent: object
    state: tuple[int, int]
    heat: int
    vertical: int
    horizontal: int
    direction: tuple[int, int]

    @property
    def state_key(self):
        return (*self.state, self.vertical, self.horizontal, self.heat)

    def __lt__(self, other):
        return self.heat < other.heat
    
    def __gt__(self, other):
        return self.heat > other.heat
    
    def __eq__(self, other):
        return self.heat == other.heat
```

```python
import heapq

initial = Point(
    parent=None,
    state=(0, 0),
    heat=0,
    vertical=0,
    horizontal=1,
    direction=tuple([0, 1]),
)

frontier = []
heapq.heappush(frontier, initial)
explored = set([initial.state_key])
width = len(cityblock[0])
height = len(cityblock)
min_heat = float('inf')

ends = []

while len(frontier) > 0:
    current = frontier.pop(0)
    row, col = current.state

    if row == height - 1 and col == width - 1:
        min_heat = min(min_heat, current.heat)
        ends.append(current)
        print(f'Found {len(ends)} paths with heat {current.heat}')

    sucessors = []
    if row + 1 < height and current.vertical < 3 and current.direction != (-1, 0):
        new_heat = cityblock[row + 1][col] + current.heat
        if new_heat < min_heat:
            if current.direction[0] == 0:
                v, h = 1, 0
            else:
                v, h = current.vertical + 1, 0
            sucessors.append(
                Point(
                    parent=current,
                    state=(row + 1, col),
                    heat=new_heat,
                    vertical=v,
                    horizontal=h,
                    direction=(1, 0),
                )
            )
    if row - 1 >= 0 and current.vertical < 3 and current.direction != (1, 0):
        new_heat = cityblock[row - 1][col] + current.heat
        if new_heat < min_heat:
            if current.direction[0] == 0:
                v, h = 1, 0
            else:
                v, h = current.vertical + 1, 0
            sucessors.append(
                Point(
                    parent=current,
                    state=(row - 1, col),
                    heat=new_heat,
                    vertical=v,
                    horizontal=h,
                    direction=(-1, 0),
                )
            )
    if col + 1 < width and current.horizontal < 3 and current.direction != (0, -1):
        new_heat = cityblock[row][col + 1] + current.heat
        if new_heat < min_heat:
            if current.direction[1] == 0:
                v, h = 0, 1
            else:
                v, h = 0, current.horizontal + 1
            sucessors.append(
                Point(
                    parent=current,
                    state=(row, col + 1),
                    heat=new_heat,
                    vertical=v,
                    horizontal=h,
                    direction=(0, 1),
                )
            )
    if col - 1 >= 0 and current.horizontal < 3 and current.direction != (0, 1):
        new_heat = cityblock[row][col - 1] + current.heat
        if new_heat < min_heat:
            if current.direction[1] == 0:
                v, h = 0, 1
            else:
                v, h = 0, current.horizontal + 1
            sucessors.append(
                Point(
                    parent=current,
                    state=(row, col - 1),
                    heat=new_heat,
                    vertical=v,
                    horizontal=h,
                    direction=(0, -1),
                )
            )

    for child in sucessors:
        if child.state_key not in explored:
            explored.add(child.state_key)
            miss_rows = height - child.state[0] - 1
            miss_cols = width - child.state[1] - 1
            if (miss_rows + miss_cols + child.heat) < min_heat:
                heapq.heappush(frontier, child)

min([e.heat for e in ends])
```

```python
import heapq

initial = Point(
    parent=None,
    state=(0, 0),
    heat=0,
    vertical=0,
    horizontal=1,
    direction=tuple([0, 1]),
)

frontier = []
heapq.heappush(frontier, initial)
explored = set([initial.state_key])
width = len(cityblock[0])
height = len(cityblock)
min_heat = float('inf')

ends = []

while len(frontier) > 0:
    current = frontier.pop(0)
    row, col = current.state

    if row == height - 1 and col == width - 1:
        min_heat = min(min_heat, current.heat)
        ends.append(current)
        print(f'Found {len(ends)} paths with heat {current.heat}')

    sucessors = []

    if row + 1 < height and current.vertical < 10 and current.direction != (-1, 0):
        new_heat = cityblock[row + 1][col] + current.heat
        if new_heat < min_heat:
            if current.direction[0] == 0:
                v, h = 1, 0
            else:
                v, h = current.vertical + 1, 0
            sucessor = Point(
                parent=current,
                state=(row + 1, col),
                heat=new_heat,
                vertical=v,
                horizontal=h,
                direction=(1, 0),
            )
            while sucessor.vertical < 4 and sucessor.state[0] + 1 < height:
                srow, scol = sucessor.state
                new_heat = cityblock[srow + 1][scol] + sucessor.heat
                sucessor = Point(
                    parent=sucessor,
                    state=(srow + 1, scol),
                    heat=new_heat,
                    vertical=sucessor.vertical + 1,
                    horizontal=sucessor.horizontal,
                    direction=(1, 0),
                )
        if new_heat < min_heat and sucessor.vertical >= 4:
            sucessors.append(sucessor)
            
    if row - 1 >= 0 and current.vertical < 10 and current.direction != (1, 0):
        new_heat = cityblock[row - 1][col] + current.heat
        if new_heat < min_heat:
            if current.direction[0] == 0:
                v, h = 1, 0
            else:
                v, h = current.vertical + 1, 0
            sucessor = Point(
                parent=current,
                state=(row - 1, col),
                heat=new_heat,
                vertical=v,
                horizontal=h,
                direction=(-1, 0),
            )
            while sucessor.vertical < 4 and sucessor.state[0] - 1 >= 0:
                srow, scol = sucessor.state
                new_heat = cityblock[srow - 1][scol] + sucessor.heat
                sucessor = Point(
                    parent=sucessor,
                    state=(srow - 1, scol),
                    heat=new_heat,
                    vertical=sucessor.vertical + 1,
                    horizontal=sucessor.horizontal,
                    direction=(-1, 0),
                )
        if new_heat < min_heat and sucessor.vertical >= 4:
            sucessors.append(sucessor)
            
    if col + 1 < width and current.horizontal < 10 and current.direction != (0, -1):
        new_heat = cityblock[row][col + 1] + current.heat
        if new_heat < min_heat:
            if current.direction[1] == 0:
                v, h = 0, 1
            else:
                v, h = 0, current.horizontal + 1
            sucessor = Point(
                parent=current,
                state=(row, col + 1),
                heat=new_heat,
                vertical=v,
                horizontal=h,
                direction=(0, 1),
            )
            while sucessor.horizontal < 4 and sucessor.state[1] + 1 < width:
                srow, scol = sucessor.state
                new_heat = cityblock[srow][scol + 1] + sucessor.heat
                sucessor = Point(
                    parent=sucessor,
                    state=(srow, scol + 1),
                    heat=new_heat,
                    vertical=sucessor.vertical,
                    horizontal=sucessor.horizontal + 1,
                    direction=(0, 1),
                )
        if new_heat < min_heat and sucessor.horizontal >= 4:
            sucessors.append(sucessor)
    if col - 1 >= 0 and current.horizontal < 10 and current.direction != (0, 1):
        new_heat = cityblock[row][col - 1] + current.heat
        if new_heat < min_heat:
            if current.direction[1] == 0:
                v, h = 0, 1
            else:
                v, h = 0, current.horizontal + 1
            sucessor = Point(
                parent=current,
                state=(row, col - 1),
                heat=new_heat,
                vertical=v,
                horizontal=h,
                direction=(0, -1),
            )
            while sucessor.horizontal < 4 and sucessor.state[1] - 1 >= 0:
                srow, scol = sucessor.state
                new_heat = cityblock[srow][scol - 1] + sucessor.heat
                sucessor = Point(
                    parent=sucessor,
                    state=(srow, scol - 1),
                    heat=new_heat,
                    vertical=sucessor.vertical,
                    horizontal=sucessor.horizontal + 1,
                    direction=(0, -1),
                )
        if new_heat < min_heat and sucessor.horizontal >= 4:
            sucessors.append(sucessor)

    for child in sucessors:
        if child.state_key not in explored:
            explored.add(child.state_key)
            miss_rows = height - child.state[0] - 1
            miss_cols = width - child.state[1] - 1
            if (miss_rows + miss_cols + child.heat) < min_heat:
                heapq.heappush(frontier, child)

min([e.heat for e in ends])
```
